```python
import pandas as pd 
import numpy as np 
import seaborn as sns 
import matplotlib.pyplot as plt 
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score,r2_score
from sklearn.compose import ColumnTransformer
from sklearn.model_selection import cross_val_score
from sklearn.linear_model import LogisticRegression

```

```python
df=pd.read_csv('World Largest Cities by Population 2024.csv')
df.drop(columns=['Unnamed: 0'])
```

```python
from scipy.stats import probplot
plt.figure(figsize=(14,4))
plt.subplot(121)
probplot(df['Population (2023)'],plot=plt)
plt.savefig('City Population In 2023.png')

plt.subplot(122)
probplot(df['Population (2024)'],plot=plt)
plt.savefig('City Population In 2024.png')
plt.show()
```
