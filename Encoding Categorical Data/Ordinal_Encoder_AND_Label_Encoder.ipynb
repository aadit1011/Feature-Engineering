```python
import pandas as pd 
import numpy as np 
import matplotlib.pyplot as plt 
import seaborn as sns 

df=pd.read_csv('customer.csv')
df 
```

```python
df=df.iloc[:,2:]
df 
```

Splitting the dataset into training and testing dataset

```python
from sklearn.model_selection import train_test_split 
x_train,x_test,y_train,y_test=train_test_split(df.drop('purchased',axis=1),df['purchased'],train_size=0.9)
```

Now Applying Ordinal Encoding

```python
from sklearn.preprocessing import OrdinalEncoder

encoder=OrdinalEncoder(categories=[['Poor', 'Average', 'Good'],['PG', 'UG', 'School']])
encoded_x_train=encoder.fit(x_train)
encoded_x_trained=encoder.transform(x_train)
encoded_x_trained=pd.DataFrame(encoded_x_trained,columns=['review','education'])
encoded_x_trained
```

```python
encoded_x_test=encoder.transform(x_test)
encoded_x_tested=pd.DataFrame(encoded_x_test,columns=['review','education'])
encoded_x_tested 
```

Applying LAbel Encoding

```python
from sklearn.preprocessing import LabelEncoder
encoder=LabelEncoder()
encoded_y_train=encoder.fit(y_train)
encoded_y_train=encoder.transform(y_train)
encoded_y_trained=pd.DataFrame(encoded_y_train,columns=['purchased'])
encoded_y_trained
```
